```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score, f1_score

```

```python
df = pd.read_csv('F:/NUPUR_PROJECT/creditcard.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
scaler = StandardScaler()  #feature scaling
df['scaled_amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df['scaled_time'] = scaler.fit_transform(df['Time'].values.reshape(-1, 1))
```

```python
df.drop(['Amount', 'Time'], axis=1, inplace=True) #drop irrevelant columns
```

```python
x=df.drop('Class',axis=1)  #create features and labels
y=df['Class']
```

```python
smote = SMOTE()  #reassemble the imbalance data
x_res, y_res = smote.fit_resample(x, y)
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_res, y_res, test_size=0.2, random_state=42)

```

```python
# Logistic Regression
log_reg = LogisticRegression(max_iter=1000)  # Increased max_iter for better convergence
log_reg.fit(x_train, y_train)

```

```python
# Predictions
y_pred_log_reg = log_reg.predict(x_test)
```

```python
#confusion matrix
CON_MATRIX = confusion_matrix(y_test, y_pred_log_reg)
print("CONFUSON MATRIX\n",CON_MATRIX)
```

```python
#accuracy calculation 
Accuracy = accuracy_score(y_test, y_pred_log_reg)
print('ACCURACY',Accuracy)
```

```python
class_report = classification_report(y_test,y_pred_log_reg)

print("Classification Report:\n", class_report)
```

```python
print("Precision:", precision_score(y_test, y_pred_log_reg))
print("Recall:", recall_score(y_test, y_pred_log_reg))
print("F1-Score:", f1_score(y_test, y_pred_log_reg))
```
